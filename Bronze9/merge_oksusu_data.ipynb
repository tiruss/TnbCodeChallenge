```python
story = concat_df["STORY"].copy()

story[0]
```

```python
import re

"""
[프리미어 월정액 고객은 별도 과금없이 시청가능 합니다.] -> 스토리와 관련이 없기 때문에 정규식을 사용해 삭제시킨다.
"""
def remove_bracket(data):
    cleaned_data = []
    
    for i in story:
        temp = re.sub('([[].*?[]])', '', i) # [] 안 내용을 삭제
        
        cleaned_data.append(temp)
    
    return cleaned_data
        
```

```python
import nltk

cleaned_data = remove_bracket(story)
tokenizer = nltk.tokenize.TweetTokenizer() # TweetTokenizer..
tokens = []

for i in cleaned_data:
    tokens.append(tokenizer.tokenize(i))
```

```python
"""
Skip_gram -> 주어진 문장에서 가능한 n개세트의 조합을 뽑아낸다.
"""
skip_grams = []
for i in tokens:
    skip_grams.append(list(nltk.skipgrams(i, 5, 3)))
    
skip_grams[0]
```

# Character 2 Vector

```python
cho = "ㄱㄲㄴㄷㄸㄹㅁㅂㅃㅅㅆㅇㅈㅉㅊㅋㅌㅍㅎ"  # len = 19
jung = "ㅏㅐㅑㅒㅓㅔㅕㅖㅗㅘㅙㅚㅛㅜㅝㅞㅟㅠㅡㅢㅣ"  # len = 21
jong = "ㄱ/ㄲ/ㄱㅅ/ㄴ/ㄴㅈ/ㄴㅎ/ㄷ/ㄹ/ㄹㄱ/ㄹㅁ/ㄹㅂ/ㄹㅅ/ㄹㅌ/ㄹㅍ/ㄹㅎ/ㅁ/ㅂ/ㅂㅅ/ㅅ/ㅆ/ㅇ/ㅈ/ㅊ/ㅋ/ㅌ/ㅍ/ㅎ".split('/')  # len = 27
test = cho + jung + ''.join(jong)

hangul_length = len(cho) + len(jung) + len(jong)  # 67
```

```python
def decompose_as_one_hot(in_char, warning=True):
    one_hot = []
    # print(ord('ㅣ'), chr(0xac00))
    # [0,66]: hangul / [67,194]: ASCII / [195,245]: hangul danja,danmo / [246,249]: special characters
    # Total 250 dimensions.
    if ord('가') <= in_char <= ord('힣'):  # 가:44032 , 힣: 55203
        x = in_char - 44032  # in_char - ord('가')
        y = x // 28
        z = x % 28
        x = y // 21
        y = y % 21
        # if there is jong, then is z > 0. So z starts from 1 index.
        zz = jong[z - 1] if z > 0 else ''
        if x >= len(cho):
            if warning:
                print('Unknown Exception: ', in_char, chr(in_char), x, y, z, zz)

        one_hot.append(x)
        one_hot.append(len(cho) + y)
        if z > 0:
            one_hot.append(len(cho) + len(jung) + (z - 1))

        return one_hot
    else:
        if in_char < 128:
            return [hangul_length + in_char]  # 67~
        elif ord('ㄱ') <= in_char <= ord('ㅣ'):
            return [hangul_length + 128 + (in_char - 12593)]  # 194~ # [ㄱ:12593]~[ㅣ:12643] (len = 51)
        elif in_char == ord('♡'):
            return [hangul_length + 128 + 51]  # 245~ # ♡
        elif in_char == ord('♥'):
            return [hangul_length + 128 + 51 + 1]  # ♥
        elif in_char == ord('★'):
            return [hangul_length + 128 + 51 + 2]  # ★
        elif in_char == ord('☆'):
            return [hangul_length + 128 + 51 + 3]  # ☆
        else:
            if warning:
                print('Unhandled character:', chr(in_char), in_char)
            return []
```

```python
decompose_as_one_hot(ord("강"))
```

```python
def decompose_str_as_one_hot(string, warning=True):
    
    temp_list = []
    for i in string:
        temp_list.extend(decompose_as_one_hot(ord(i), warning=warning))

    return list(temp_list)
```
